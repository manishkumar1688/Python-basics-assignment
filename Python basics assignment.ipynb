**Q7. What is the difference between "==" and "is" operators in python?**

  -   '==' this is an equality operator which compares the values of two objects. For example, if you compare two dog with the == operator, it will return "both dogs are equal" if they have the same contents.

  -  'is' this is an identity operator which compares if two variables point to the same object in memory. For example, if you compare two Dogs with the 'is' operator, it will return "these are two different dogs" if they are not the same object.

  For Ex:

```python
# 'is' operaator
a = [5,6,9,10]
b = [2,3,4,5,6]
c = a
```

```python
if a is b:  # 'is' - it is pointing at two seperate variable object in memory.
  print(True)
else:
  print(False)
```

```python
if a is c:   # 'is' - it is pointing at two seperate variable object in memory.
  print(True)
else:
  print(False)
```

```python
# '==' equality operator
a = [5,6,9,10]
b = [5,6,9,10]
c = a
if a == b:  # equality operator is checking whether contents in both the variables are equal to each other or not.
  print(True)
else:
  print(False)
```

```python
a = [5,6,9,10]
b = [5,6,9,11]
c = a
if a == b:   # equality operator is checking whether contents in both the variables are equal to each other or not.
  print(True)
else:
  print(False)
```

**Q14. Describe a scenario where a while loop is more suitable than a for loop**

  -    while loops are best suited when you do not know

 -     Scenario -
                Find the square of the each no. till 50 but, if square of the number should not be grater than 40.
               

```python
n = 50
def square_no(n):
  for i in n:
    print(i**2)
```
